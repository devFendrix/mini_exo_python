# Comment varie les salaires dans la data dans le monde ?
##### Importation des librairies requises

```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sb
import numpy as np

salaries = pd.read_csv('data/ds_salaries.csv', sep=',')
salaries.head(500)
```

### Analyse des salaires français en fonction du niveau d'expérience
EN = Débutant / MI = Junior / SE = Senior / EX = Expert

```python
sb.set_theme()
salariesFR = salaries.query('company_location == "FR"')
sb.relplot(data = salariesFR, x="experience_level", y = "salary")
```

```python
sb.set_theme()
sb.relplot(data = salaries, x="experience_level", y = "salary")
```

```python
mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data = salariesFR, order = salariesFR['job_title'].value_counts().index)
mpl.xticks(rotation=90)
```

```python
mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data = salaries, order = salaries['job_title'].value_counts().index)
mpl.xticks(rotation=90)
```

```python
top_salarie = salariesFR.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).head(20)
top_salarie = pd.DataFrame(top_salarie).reset_index()

mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data=salariesFR, order = salaries.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).iloc[:20].index)
mpl.xticks(rotation=90)
```

```python
top_salarie = salaries.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).head(20)
top_salarie = pd.DataFrame(top_salarie).reset_index()

mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data=salaries, order = salaries.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).iloc[:20].index)
mpl.xticks(rotation=90)
```

```python
countFR = len(salaries.query('company_location == "FR"'))
moyenneFR = salariesFR['salary'].sum() / countFR

countUS = len(salaries.query('company_location == "US"'))
moyenneUS = salaries.query('company_location == "US"')['salary'].sum() / countUS

countGB = len(salaries.query('company_location == "GB"'))
moyenneGB = salaries.query('company_location == "GB"')['salary'].sum() / countGB

countES = len(salaries.query('company_location == "ES"'))
moyenneES = salaries.query('company_location == "ES"')['salary'].sum() / countES

print(moyenneFR)
print(moyenneUS * 1.02)
print(moyenneGB * 1.14)
seg = pd.DataFrame(data={"origine": ["FR", "US", "GB", "ES"], "Salaires": [moyenneFR, moyenneUS * 1.02, moyenneGB * 1.14, moyenneES]})
sb.barplot(data=seg, x="origine", y="Salaires")

```

```python
salariesFRYears = dict()
salariesUSYears = dict()
salariesGBYears = dict()
salariesESYears = dict()

salariesFRYears['2020'] = salaries.query('company_location == "FR" & work_year == 2020')['salary'].sum() / countFR
salariesUSYears['2020'] = salaries.query('company_location == "US" & work_year == 2020')['salary'].sum() / countUS
salariesGBYears['2020'] = salaries.query('company_location == "GB" & work_year == 2020')['salary'].sum() / countGB
salariesESYears['2020'] = salaries.query('company_location == "ES" & work_year == 2020')['salary'].sum() / countES
salariesFRYears['2021'] = salaries.query('company_location == "FR" & work_year == 2021')['salary'].sum() / countFR
salariesUSYears['2021'] = salaries.query('company_location == "US" & work_year == 2021')['salary'].sum() / countUS
salariesGBYears['2021'] = salaries.query('company_location == "GB" & work_year == 2021')['salary'].sum() / countGB
salariesESYears['2021'] = salaries.query('company_location == "ES" & work_year == 2021')['salary'].sum() / countES
salariesFRYears['2022'] = salaries.query('company_location == "FR" & work_year == 2022')['salary'].sum() / countFR
salariesUSYears['2022'] = salaries.query('company_location == "US" & work_year == 2022')['salary'].sum() / countUS
salariesGBYears['2022'] = salaries.query('company_location == "GB" & work_year == 2022')['salary'].sum() / countGB
salariesESYears['2022'] = salaries.query('company_location == "ES" & work_year == 2022')['salary'].sum() / countES

salariesFRYearsDf = pd.DataFrame.from_dict(salariesFRYears, orient='index')
salariesUSYearsDf = pd.DataFrame.from_dict(salariesUSYears, orient='index')
salariesGBYearsDf = pd.DataFrame.from_dict(salariesGBYears, orient='index')
salariesESYearsDf = pd.DataFrame.from_dict(salariesESYears, orient='index')
salariesFRYearsDf.index.names = ['Année']
salariesUSYearsDf.index.names = ['Année']
salariesGBYearsDf.index.names = ['Année']
salariesESYearsDf.index.names = ['Année']
salariesFRYearsDf = salariesFRYearsDf.rename(columns = {0:'Salaire'})
salariesUSYearsDf = salariesUSYearsDf.rename(columns = {0:'Salaire'})
salariesGBYearsDf = salariesGBYearsDf.rename(columns = {0:'Salaire'})
salariesESYearsDf = salariesESYearsDf.rename(columns = {0:'Salaire'})
salariesFRYearsDf['Pays'] = "Français"
salariesUSYearsDf['Pays'] = "Américains"
salariesGBYearsDf['Pays'] = "Anglais"
salariesESYearsDf['Pays'] = "Espagnol"

sb.lineplot(data=salariesFRYearsDf, x='Année', y='Salaire', hue='Pays', palette=['blue'])
sb.lineplot(data=salariesUSYearsDf, x='Année', y='Salaire', hue='Pays', palette=['black'])
sb.lineplot(data=salariesGBYearsDf, x='Année', y='Salaire', hue='Pays', palette=['red'])
sb.lineplot(data=salariesESYearsDf, x='Année', y='Salaire', hue='Pays', palette=['green'])
```
