# Comment varie les salaires dans la data dans le monde ?
##### Importation des librairies requises

```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sb
import numpy as np

salaries = pd.read_csv('data/ds_salaries.csv', sep=',')
salaries.head(500)
```

### Analyse des salaires français en fonction du niveau d'expérience
EN = Débutant / MI = Junior / SE = Senior / EX = Expert

```python
sb.set_theme()
salariesFR = salaries.query('company_location == "FR"')
sb.relplot(data = salariesFR, x="experience_level", y = "salary")
```

```python
sb.set_theme()
sb.relplot(data = salaries, x="experience_level", y = "salary")
```

```python
mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data = salariesFR, order = salariesFR['job_title'].value_counts().index)
mpl.xticks(rotation=90)
```

```python
mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data = salaries, order = salaries['job_title'].value_counts().index)
mpl.xticks(rotation=90)
```

```python
top_salarie = salariesFR.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).head(20)
top_salarie = pd.DataFrame(top_salarie).reset_index()

mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data=salariesFR, order = salaries.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).iloc[:20].index)
mpl.xticks(rotation=90)
```

```python
top_salarie = salaries.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).head(20)
top_salarie = pd.DataFrame(top_salarie).reset_index()

mpl.figure(figsize=(15, 10))
sb.countplot(x="job_title", data=salaries, order = salaries.groupby(by=['job_title'])['work_year'].count().sort_values(ascending=False).iloc[:20].index)
mpl.xticks(rotation=90)
```
